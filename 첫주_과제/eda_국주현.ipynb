```python
import pandas as pd
df = pd.read_csv('/content/drive/MyDrive/투빅스/HR_Employee_Attrition.csv')
df
```

```python
df.isnull().sum() #결측치 개수 확인
```

```python
NUMS = []
for col in df.columns:
  if df[col].dtypes=='int64':
    # print(col)
    NUMS.append(col)

df_nums = df[NUMS]
df_nums
```

```python
#이상치 제거하자
q3=df_nums.quantile(0.75)
q1=df_nums.quantile(0.25)
iqr=q3-q1
iqr
```

```python
for j in NUMS:
  def is_outlier(df):
    score=df[j]
    if score > q3[j] + 1.5*iqr[j] or score < q1[j]-1.5*iqr[j]:
      return True
    else:
      return False
  df_nums[j+'_이상치여부']=df_nums.apply(is_outlier, axis=1)
df_nums
```

```python
for j in NUMS:
  df_nooutlier=df_nums.loc[df_nums[j+'_이상치여부']==False]
for j in NUMS:
  del df_nooutlier[j+'_이상치여부']
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
import matplotlib as mpl
```

```python
print('Attrition 비율')
print(f'{df.Attrition.value_counts().index[0]} : {round(df.Attrition.value_counts()[0] / len(df), 2) * 100}%')
print(f'{df.Attrition.value_counts().index[1]} : {round(df.Attrition.value_counts()[1] / len(df), 2) * 100}%')
```

```python
plt.figure(figsize = (12,12))
sns.countplot(x = df['Attrition'])
plt.title('Attrition의 분포')
plt.show()
```

```python
category= []
for column in df.columns:
    if df[column].dtype == object or len(df[column].value_counts())<10: #같은 값이 20개 이상 있으면 범주형 취급
        category.append(column)
```

```python
len(category)
```

```python
fig, ax = plt.subplots(7, 3, figsize=(28, 28), constrained_layout=True)
ax = ax.flatten()
fig.suptitle('Attrition과 범주형 분포', fontsize=16)

for i in range(len(category)):
    sns.countplot(x=category[i], data=df,
                      hue='Attrition', ax=ax[i]).set(xlabel = None)
    ax[i].set(title = category[i])
    
    if df[category[1]].nunique() >= 3:
        ax[i].tick_params(labelrotation=20)
        
plt.show()
```

```python
#연속형 컬럼 확인
# continuous column
cont_cols = []
for column in df.columns:
    if column not in category:
        cont_cols.append(column)
```

```python
cont_cols
```

```python
fig, ax = plt.subplots(4, 4, figsize=(28, 14), constrained_layout=True)
ax = ax.flatten()
fig.suptitle('연속형 Column들의 분포', fontsize=16)

for i in range(len(cont_cols)):
    sns.distplot(df[df['Attrition'] == 'Yes'][cont_cols[i]], color='Green', ax=ax[i], hist = False).set(xlabel = None, ylabel = None)
    sns.distplot(df[df['Attrition'] == 'No'][cont_cols[i]], color='Yellow', ax=ax[i],hist = False).set(xlabel = None, ylabel = None)
    ax[i].set(title = cont_cols[i])
    
plt.show()
```

```python
# boxplot
fig, ax = plt.subplots(4, 4, figsize=(28, 14), constrained_layout=True)
ax = ax.flatten()
fig.suptitle('박스플롯', fontsize=16)

for i in range(len(cont_cols)):
    ax[i].set(title = cont_cols[i])
    sns.boxplot(x=df['Attrition'], y = df[cont_cols[i]], ax=ax[i]).set(xlabel=None,  ylabel=None)
    
plt.show()
```

```python
def age(df):
  a=df['Age']
  if a >= 18 and a < 20:
    return 1
  elif a >= 20 and a < 30:
    return 2
  elif a >= 30 and a < 40:
    return 3
  elif a >= 40 and a < 50:
    return 4
  elif a >= 50 and a < 60:
    return 5
  else:
    return 6

df['연령대']=df.apply(age, axis=1)
df
```
